```python
#house price dataset
house = pd.read_csv('BostonHousing.csv')
```

```python
house.head()
```

```python
 sns.distplot(house['medv'])
#distplot
```

```python
#correlation 
# +ve correlation
# -ve correlation
correlation = house.corr()
```

```python
#using heatmap
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square =True,fmt='.1f',annot=True,annot_kws={'size':10},cmap='Blues')
# cbar = color bar , square ,fmt = how many points after . , annot = annotation,annot_kws = for defining size of annotation,cmap = color of correlation map
```
