```python
import matplotlib.pyplot as plt #importing matplotlib
import numpy as np
%matplotlib inline
plt.show()
```

```python
x = np.linspace(0,10,100)  #it will give us 100 evenly spaced values between 0 and 10
y = np.sin(x) #x is from upper values
z = np.cos(x) 
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#sin wave
plt.plot(x,y)
```

```python
#cos wave
plt.plot(x,z)
```

```python
#adding titles to graph
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin values')
plt.title('sin wave')
```

```python
#parabola
x = np.linspace(0,10,100)
y = x**3
plt.plot(x,y)
plt.xlabel('spaced values')
plt.ylabel('curve')
plt.title('parabola')
```

```python
#different pattern in graphs
plt.plot(x,y,'r+') #r = red and + is used as pattern , put inside ''
```

```python
plt.plot(x,y,'g*')
```

```python
plt.plot(x,y,'b.')
```

```python
plt.plot(x,y,'yx')
#more in matplotlib documentation
```

```python
#multiple graphs in one
x = np.linspace(5,50,100)
plt.plot(x,np.sin(x),'r-') #declaring sin(x) in single line only no need to define it in seperate varibale
plt.plot(x,np.cos(x),'y--')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) #0,0 = origin and 1,1 = height, width respectively . this will be enclosed in a rectangle that's why we have to mention it
languages = ['english','french','hindi','spanish']
people = [100,20,70,10]
plt.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('PEOPLE')
#bar-graph
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['english','french','hindi','spanish']
people = [100,20,70,10]
ax.pie(people,labels =languages,autopct='%.1f%%') #autopct = '%.1f%%' # display the percentage value to 1 decimal place
#autopct = '%.2f%%' # display the percentage value to 2 decimal places
plt.show()
#pie-chart
```

```python
x = np.linspace(10,20,30)
y = np.sin(x)
z = np.cos(x)
fig4 = plt.figure()
ax = fig4.add_axes([0,0,1,1])
ax.scatter(x,y,color = 'b')
ax.scatter(x,z,color = 'y')
plt.show()
#scatter-plot
```

```python
fig5 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(900)
y = np.sin(z)
x = np.sin(z)
ax.scatter(x,y,z, c = z,cmap = 'Blues')
#3d scatter plot
```
