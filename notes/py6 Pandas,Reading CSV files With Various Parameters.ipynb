# #Quoting and Escape Characters

```python
#if there is a specific character inside a csv file then we can escape it with inbuilt function.
```

```python
data = 'a,b\n"hello, \\"zera\\", nice to see you",5'
```

```python
pd.read_csv(StringIO(data),escapechar="\\")
```
